# 结巴分词实践

## 结巴分词的三种模式

### 精确模式

```python
import jieba
s = "我来到北京清华大学"
result = jieba.cut(s, cut_all=False)
print("精确模式：", " / ".join(result))
```

### 全模式

```python
s = "我来到北京清华大学"
result = jieba.cut(s, cut_all=True)
print("全模式：", " / ".join(result))
```

### 搜索引擎模式

```python
s = "小明硕士毕业于中国科学院计算所，后在日本京都大学深造"
result = jieba.cut_for_search(s)
print("搜索引擎模式：", " / ".join(result))
```

## 载入自定义词典

```python
jieba.load_userdict("userdict.txt")

s = "hjy即将是自然语言处理方面的高手。"
result = jieba.cut(s, HMM=False)
print(" / ".join(result))
```
