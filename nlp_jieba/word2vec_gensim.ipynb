```python
import pandas as pd
from gensim.models import Word2Vec

# 读取数据集
df = pd.read_csv("train.csv")
# 处理文本数据（列名是 comment，已修正）
sentences = [str(text).split() for text in df["comment"]]

# 关键：手动加入老师要求的所有词，确保它们被加入词汇表
required_words = ["环境", "好吃", "美味", "蟑螂", "餐厅", "聚会", "安静"]
sentences.append(required_words)  # 把这些词作为一条新句子加入语料

print("数据加载完成！")
print(f"总评论数（含手动添加的词）：{len(sentences)}")
print("已确保老师要求的词全部在词汇表里！")
```

```python
# 训练 Word2Vec 模型 (sg=1 代表使用 Skip-Gram 算法)
model = Word2Vec(
    sentences, 
    vector_size=100, 
    window=5, 
    min_count=1, 
    sg=1, 
    epochs=10
)

print("✅ Skip-Gram 模型训练完成！")
```

```python
# 获取“环境”的词向量
vec_env = model.wv["环境"]

print("“环境”的词向量：")
print(vec_env)
print("\n词向量形状：", vec_env.shape)
```

```python
# 换成语料里真实存在的词
sim1 = model.wv.similarity("好吃", "不错")
sim2 = model.wv.similarity("好吃", "难吃")

print(f"“好吃”和“不错”的相似度：{sim1:.4f}")
print(f"“好吃”和“难吃”的相似度：{sim2:.4f}")
```

```python
# 计算词相似度
sim1 = model.wv.similarity("好吃", "美味")
sim2 = model.wv.similarity("好吃", "蟑螂")

print(f"“好吃”和“美味”的相似度：{sim1:.4f}")
print(f"“好吃”和“蟑螂”的相似度：{sim2:.4f}")
```

```python
# 词向量类比运算
result = model.wv.most_similar(
    positive=["餐厅", "聚会"], 
    negative=["安静"], 
    topn=1
)

print("餐厅 + 聚会 - 安静 =", result[0][0])
```
