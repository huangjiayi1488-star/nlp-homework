```python
import pandas as pd
import jieba
from gensim.models import Word2Vec

# 1. 读取数据
df = pd.read_csv("train.csv")
sentences = []

# 2. 分词（用正确的列名 comment）
for text in df["comment"]:
    words = jieba.lcut(str(text))
    sentences.append(words)

# 3. 用 Skip-Gram 训练 Word2Vec（任务四核心要求）
model = Word2Vec(sentences, vector_size=100, window=5, min_count=2, sg=1, workers=3)

# 4. 输出“环境”的词向量
print("向量维度:", model.wv["环境"].shape)
print("环境向量:", model.wv["环境"])

# 5. 输出和“好吃”最相似的3个词
print("\n与好吃相似的词：")
print(model.wv.most_similar("好吃", topn=3))

# 6. 计算相似度
print("\n相似度计算：")
print("好吃 vs 美味:", model.wv.similarity("好吃", "美味"))
print("好吃 vs 蟑螂:", model.wv.similarity("好吃", "蟑螂"))

# 7. 向量运算：餐厅 + 聚会 - 安静
print("\n向量运算：餐厅 + 聚会 - 安静")
print("与「餐厅 + 聚会 - 安静」最相似的词：")
print(model.wv.most_similar(positive=["餐厅", "聚会"], negative=["安静"], topn=1))
```
